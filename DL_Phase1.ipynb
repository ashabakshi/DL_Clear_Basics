# 🧠 Phase 1 — What is Deep Learning?

```python
# First, the big picture
# AI > Machine Learning > Deep Learning

# AI — machines doing smart things
# ML — machines learning from data
# Deep Learning — ML but using Neural Networks (inspired by human brain)

# Think of it like this:

# Normal ML = you give rules + data → machine predicts
# Deep Learning = you just give data → machine figures out the rules itself
```

```python
# Why "Deep"?
# Because there are multiple hidden layers — the more layers, the "deeper" the network.
# Input → [Layer 1] → [Layer 2] → [Layer 3] → Output
#               ↑ these are the "deep" layers
# Each layer learns something more complex than the previous one.
# Example — recognizing a cat photo:

# Layer 1 → learns edges
# Layer 2 → learns shapes (ears, eyes)
# Layer 3 → learns "oh this is a cat!"
```

```python
# One Neuron — How it works
# A single neuron does just 3 things:
# Step 1 — Take inputs & multiply by weights
# Step 2 — Add bias
# Step 3 — Pass through activation function
# output = activation( (x1×w1) + (x2×w2) + bias )
```

```python

import numpy as np

x = np.array([2, 3])        # inputs
w = np.array([0.5, 0.8])    # weights
b = 1                        # bias

z = np.dot(x, w) + b        # step 1 & 2
print(z)                     # 4.4

output = max(0, z)           # step 3 — ReLU activation
print(output)                # 4.4
```

```python
# Activation Functions — why needed?

# Without activation function, no matter how many layers you add, the network behaves like a single straight line — it can't learn complex patterns.
# Activation function adds curves → network learns complex things.

# 3 most important ones:
```

```python
import numpy as np

# ReLU — most common, used in hidden layers
def relu(z):
    return np.maximum(0, z)
# negatives → 0, positives → same

# Sigmoid — output between 0 and 1
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
# used in binary classification (yes/no problems) ->0-1

# Softmax — for multiple classes
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))
# used in output layer when classes > 2
```

```python
# Full picture — data flowing through a network
# Input Data
#     ↓
# [Neuron: multiply weights, add bias, apply ReLU]
#     ↓
# [Neuron: multiply weights, add bias, apply ReLU]
#     ↓
# [Output Neuron: apply Sigmoid or Softmax]
#     ↓
# Final Prediction
# This flow — input to output — is called the Forward Pass.
```

```python
import numpy as np

def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Input
x = np.array([2.0, 3.0])

# Hidden layer — 3 neurons
W1 = np.array([[0.5, 0.8],
                [0.2, 0.9],
                [0.1, 0.4]])
b1 = np.array([1.0, 0.5, 0.3])

hidden = relu(np.dot(W1, x) + b1)
print("Hidden layer output:", hidden)

# Output layer — 1 neuron (binary classification)
W2 = np.array([0.6, 0.3, 0.7])
b2 = 0.2

output = sigmoid(np.dot(W2, hidden) + b2)
print("Final prediction:", output)
# output close to 1 = class 1, close to 0 = class 0
```
